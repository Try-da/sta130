Proposal 1: The Impact of Fitness on Weight Loss

1. Statement of Research Question:
How do fitness activities affect weight loss outcomes? Specifically, does regular participation in different types and frequencies of fitness activities significantly promote the reduction of body weight and body fat?
2. Variables to Explore and How to Explore Them:
a. Independent Variables:
 Fitness Frequency (times per week): Chosen because the frequency directly influences calorie expenditure and metabolism. Increasing frequency may enhance weight loss effects.
 Fitness Type (aerobic, anaerobic, mixed): Different exercise types affect fat burning and muscle growth differently. Selecting this variable helps determine which type is more effective.
b. Dependent Variables:
 Change in Body Weight (kg): Directly measures the effect of weight loss.
 Change in Body Fat Percentage (%): Provides a more accurate reflection of body composition changes, focusing on fat reduction.
c. Control Variables:
 Dietary Habits: Diet significantly impacts weight loss; controlling for this eliminates confounding effects.
 Age and Gender: Physiological differences may affect weight loss outcomes.
 Basal Metabolic Rate: Individual metabolic rates vary, influencing the speed of weight loss.
d. Why These Variables Were Chosen:
 Fitness Frequency and Type: These are key adjustable factors in a fitness regimen that directly impact weight loss outcomes.
 Body Weight and Fat Percentage Changes: Core indicators for assessing the success of weight loss efforts.
 Control Variables: To ensure result reliability by eliminating other factors that might influence weight loss.
e. Exploration Methods:
 Data Collection: Track a group of participants attempting to lose weight, recording their fitness frequency, type, and changes in body metrics.
f. Visualization:
 Line Charts: Show trends in weight and body fat percentage over time to observe long-term effects.
 Scatter Plots: Illustrate the relationship between fitness frequency/type and changes in weight/body fat percentage to identify correlations.
g. Reasons for Visualization Choices:
 Line Charts: Suitable for displaying trends in time-series data, intuitively presenting the weight loss process.
 Scatter Plots: Help discover relationships between independent and dependent variables, identifying correlations or patterns.
3. Planned Analysis:
a. Methods:
 Linear Regression Analysis: Evaluate the impact of fitness frequency and type on weight loss outcomes.
 Multiple Regression Analysis: Incorporate control variables for a more precise assessment of independent variables.
b. Assumptions:
 Linear Relationship: Assuming a linear relationship between fitness frequency/type and weight loss outcomes.
 Independence of Observations: Each participant’s data is independent of others.
 Normality of Residuals: Residuals are normally distributed.
 Homoscedasticity: Variance of residuals is consistent across all levels of independent variables.
4. Hypothesized Results and Their Impact:
a. Expected Results:
 Higher fitness frequency and specific types (e.g., aerobic exercises) are significantly associated with greater weight loss.
b. Implications:
 Confirms the positive effect of fitness on weight loss, guiding individuals to develop effective workout plans.
 Provides scientific evidence to help fitness trainers and healthcare professionals offer personalized weight loss advice.


Proposal 2: The Relationship Between Education Level and Salary

1. Statement of Research Question:
How does education level affect salary? Specifically, does a higher education level lead to higher salary levels? Are there other influencing factors to consider?
2. Variables to Explore and How to Explore Them:
a. Independent Variable:
 Education Level (high school, associate, bachelor’s, master’s, doctorate): Education is often viewed as a critical factor influencing employment and salary.
b. Dependent Variable:
 Salary Level (monthly or annual): Directly reflects economic benefits.
c. Control Variables:
 Work Experience: Experience may increase salary; controlling for it is necessary.
 Industry Type: Significant salary differences exist across industries.
 City Tier: Regional economic disparities affect salary levels.
 Age and Gender: Potential salary disparities need consideration.
d. Why These Variables Were Chosen:
 Education Level: Exploring the return on educational investment is significant for individuals and society.
 Salary Level: Directly measures economic outcomes.
 Control Variables: Eliminating other factors that may affect salary ensures analysis accuracy.
e. Exploration Methods:
 Data Collection: Gather salary data from individuals with various education levels, along with control variables.
f. Visualization:
 Box Plots: Compare salary distributions across different education levels, observing medians and variability.
 Scatter Plots: Display the relationship between education level and salary, identifying trends and outliers.
g. Reasons for Visualization Choices:
 Box Plots: Effectively display data distribution characteristics, suitable for multi-group comparisons.
 Scatter Plots: Visually show relationships between two continuous variables, aiding in pattern discovery.
3. Planned Analysis:
a. Methods:
 One-Way ANOVA: Test for significant differences in mean salaries among different education levels.
 Multiple Regression Analysis: Control for other variables to assess the independent impact of education on salary.
b. Assumptions:
 Normality: Salary data is approximately normally distributed within each education level.
 Homogeneity of Variance: Salary variances are equal across groups.
 Independence of Observations: Each individual’s data is independent.
4. Hypothesized Results and Their Impact:
a. Expected Results:
 Higher education levels are significantly associated with higher salaries, indicating noticeable salary increases with each educational advancement.
b. Implications:
 Highlights the importance of education for personal economic development, supporting policies that encourage continued education.
 Assists individuals in making informed decisions about education and career paths.

Proposal 3: The Relationship Between Electronic Device Usage and Health

1. Statement of Research Question:
How does frequent use of electronic devices affect health status? Specifically, is there a correlation between daily electronic device usage time and health indicators like vision, sleep quality, and frequency of neck pain?
2. Variables to Explore and How to Explore Them:
a. Independent Variables:
 Daily Electronic Device Usage Time (hours): Usage duration directly affects exposure level.
 Usage Purpose (work, entertainment, study): Different purposes may lead to varying usage patterns and postures, affecting health.
b. Dependent Variables:
 Vision Status: Degree of myopia, symptoms of eye strain.
 Sleep Quality: Time to fall asleep, sleep duration, number of sleep interruptions.
 Frequency of Neck Pain: Number of occurrences within a specified period.
c. Control Variables:
 Age and Gender: Physiological differences may influence health indicators.
 Exercise Habits: Regular exercise may mitigate negative effects of prolonged device use.
 Work Environment: Availability of ergonomic equipment.
d. Why These Variables Were Chosen:
 Electronic Device Usage Time: A key indicator for assessing the impact on health.
 Health Indicators: Common health issues related to device usage, facilitating impact evaluation.
 Control Variables: Improves analysis accuracy by eliminating other health-influencing factors.
e. Exploration Methods:
 Data Collection: Use surveys or logs to collect data on participants’ device usage and health status.
f. Visualization:
 Correlation Matrix: Show correlations between independent and dependent variables to identify strong relationships.
 Grouped Bar Charts: Compare health problem incidences across different usage durations.
g. Reasons for Visualization Choices:
 Correlation Matrix: Allows simultaneous observation of multiple variable correlations, revealing potential patterns.
 Grouped Bar Charts: Visually compares differences between groups, suitable for categorical data.
3. Planned Analysis:
a. Methods:
 Pearson Correlation Analysis: Evaluate linear correlations between continuous variables.
 Multiple Regression Analysis: Control for other variables to assess the independent impact of device usage time on health indicators.
b. Assumptions:
 Linear Relationship: Assuming a linear relationship between independent and dependent variables.
 Normality: Health indicator data is approximately normally distributed.
 Independence of Observations: Each participant’s data is independent.
4. Hypothesized Results and Their Impact:
a. Expected Results:
 Longer device usage is associated with poorer vision, lower sleep quality, and higher frequency of neck pain.
 Usage purpose affects health differently; prolonged entertainment use may pose greater health risks than work-related use.
b. Implications:
 Encourages the public to reasonably limit device usage time and pay attention to posture and environment.
 Provides a basis for developing health guidelines and policies, promoting healthier lifestyles.
